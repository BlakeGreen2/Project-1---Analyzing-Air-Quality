```python
import pandas as pd
```

```python
air_data = pd.read_csv('AirQuality_Daily_StudentVersion(in).csv')

air_data = pd.DataFrame(air_data)

#air_data

air_data.head(3)
```

Client Request 1 is to find the 5 locations in Nebraska with the highest mean and median concentrations of voc, PM 2.5, and PM 10.0

```python
data_grouped_location = air_data.groupby('sensor.name')


voc_locations = data_grouped_location.agg(mean_voc = ('voc', 'mean'),
                                               median_voc = ('voc','median'))
PM2_5_locations = data_grouped_location.agg(mean_PM2_5 = ('pm2.5_alt', 'mean'),
                                               median_PM2_5 = ('pm2.5_alt','median'))
PM10_locations = data_grouped_location.agg(mean_PM10 = ('pm10.0_atm', 'mean'),
                                               median_PM10 = ('pm10.0_atm','median'))


high_PM2_5_locations = PM2_5_locations.sort_values(by = ['mean_PM2_5', 'median_PM2_5'], ascending = [True,True])
high_PM10_locations = PM10_locations.sort_values(by = ['mean_PM10', 'median_PM10'], ascending = [True,True])
high_voc_locations = voc_locations.sort_values(by = ['mean_voc', 'median_voc'], ascending = [True,True])

high_voc_locations.tail(5) 
```

Client request 3 is to find out if humidity and temperature have noticeable effects on air quality.

```python
def assign_temperature(temperature):
    if temperature < 32:
        return 'Below Freezing'
    elif 32 <= temperature <= 50:
        return 'Cool'
    elif 51 <= temperature <= 70:
        return 'Warm'
    else:
        return 'Hot'

air_data['temperature_level'] = air_data['temperature'].apply(assign_temperature)

grouped_temps = air_data.groupby('temperature_level')

temp_data = grouped_temps.agg(mean_voc = ('voc', 'mean'),
                              median_voc = ('voc', 'median'),
                              mean_pm10 = ('pm10.0_atm', 'mean'),
                              median_pm10 = ('pm10.0_atm', 'median'),
                              mean_pm2_5 = ('pm2.5_alt', 'mean'),
                              median_pm2_5 = ('pm2.5_alt', 'median'))
temp_data
```

```python
def assign_humidity(humidity):
    if humidity < 50:
        return 'Low'
    elif 50 <= humidity < 80:
        return 'High'
    else:
        return 'Very High'

air_data['humidity_level'] = air_data['humidity'].apply(assign_humidity)

grouped_humidities = air_data.groupby('humidity_level')

humidity_data = grouped_humidities.agg(mean_voc = ('voc', 'mean'),
                              median_voc = ('voc', 'median'),
                              mean_pm10 = ('pm10.0_atm', 'mean'),
                              median_pm10 = ('pm10.0_atm', 'median'),
                              mean_pm2_5 = ('pm2.5_alt', 'mean'),
                              median_pm2_5 = ('pm2.5_alt', 'median'))
humidity_data
```

Client request 4 is to analyze the data to see if there have been any AQI health risks for sensitive populations based on EPA AQI ratings.

```python
def assign_unhealthy_sensitive(pm2_5):
    if pm2_5 <= 35.4:
        return 'Safe'
    if 35.5 <= pm2_5 <=55.4:
        return 'Unhealthy for Sensitive Groups'
    else:
        return 'Unsafe'

air_data['pm2_5_category'] = air_data['pm2.5_alt'].apply(assign_unhealthy_sensitive)

sensitive_only = air_data[air_data['pm2_5_category'] == 'Unhealthy for Sensitive Groups']

sensitive_only
```

```python
def assign_unhealthy_sensitive2(pm_10):
    if pm_10 < 154:
        return 'Safe'
    elif 155 <= pm_10 <= 254:
        return 'Unhealthy for Sensitive Groups'
    else:
        return 'Unsafe'

air_data['pm_10_category'] = air_data['pm10.0_atm'].apply(assign_unhealthy_sensitive2)

sensitive_only2 = air_data[air_data['pm_10_category'] == 'Unhealthy for Sensitive Groups']

sensitive_only2
```

Client request 5 is to find out analyze the data to see if sensor altitude has an impact on air quality values.

```python
def assign_altitude(altitude):
    if altitude < 1600:
        return 'Low'
    elif 1600 <= altitude < 2600:
        return 'Medium'
    else:
        return 'High'

air_data['altitude_category'] = air_data['sensor.altitude'].apply(assign_altitude)

grouped_altitude = air_data.groupby('altitude_category').agg(mean_voc = ('voc', 'mean'),
                              median_voc = ('voc', 'median'),
                              mean_pm10 = ('pm10.0_atm', 'mean'),
                              median_pm10 = ('pm10.0_atm', 'median'),
                              mean_pm2_5 = ('pm2.5_alt', 'mean'),
                              median_pm2_5 = ('pm2.5_alt', 'median'))
grouped_altitude
```
